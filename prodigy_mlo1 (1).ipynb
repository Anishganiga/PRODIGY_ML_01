```python
import pandas as pd
data=pd.read_csv('C:/Users/Anish/Documents/house prices.csv')
data


```

```python
print(data.columns)
```

```python
print(data.describe())
```

```python
from sklearn.model_selection import train_test_split
features=['GrLivArea','FullBath', 'BedroomAbvGr']
X=data[features]
y=data['SalePrice']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression 
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
print("Mean squared error =",mse)
r2=r2_score(y_test,y_pred)
print("R squared =",r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual prices")
plt.ylabel("Predicted Prices")
plt.title("Actual prices vs predicted prices")
plt.show()
```

```python
nd=pd.DataFrame({'GrLivArea':[3000],'FullBath':[3], 'BedroomAbvGr':[5]})
pp=model.predict(nd)
print(f"The predicted price of the House is {pp[0]}")
```
